```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('/Users/manasas/Desktop/Manasa.jpeg')
```

```python
image = cv2.imread('/Users/manasas/Desktop/Manasa.jpeg')
# Convert the image from BGR to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb)
plt.axis('off')
plt.show()
```

```python
blurred_image = cv2.GaussianBlur(image, (9, 9), 0)
image_rgb1 = cv2.cvtColor(blurred_image, cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb1)
plt.axis('off')
plt.show()

```

```python
sharpened_image = cv2.addWeighted(image, 1.5, image, -0.5, 0)

image_rgb2 = cv2.cvtColor(sharpened_image, cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb2)
plt.axis('off')
plt.show()
```

```python
edges = cv2.Canny(image, threshold1=100, threshold2=200)

image_rgb3 = cv2.cvtColor(edges, cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb3)
plt.axis('off')
plt.show()
```
